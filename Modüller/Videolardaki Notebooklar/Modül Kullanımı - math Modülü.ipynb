```python
def factorial(sayı):
    print("Benim fonksiyonum")
    
    faktoriyel = 1
    if (sayı == 0 or sayı == 1):
        return 1
    else:
        while (sayı >= 1):
            faktoriyel *= sayı
            sayı -= 1
        return faktoriyel
```
